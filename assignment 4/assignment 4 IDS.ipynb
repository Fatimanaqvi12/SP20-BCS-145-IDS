```python
#Print the percentage of students who have a CGPA of 3.0 or above.
count = 0
for item in df.CGPA:
       if (float(item) >= 3):
        count += 1
print("CGPA of 3.0 or above:", count)
print("Percentage: ",(count/len(df.CGPA))*100)
```
